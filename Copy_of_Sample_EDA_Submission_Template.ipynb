## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the original dataset

df1 = df.copy()
````

````python
# Check missing values

df1.isnull().sum()
````

````python
# Remove duplicate rows

df1.drop_duplicates(inplace=True)

print("Dataset Shape After Removing Duplicates:", df1.shape)
````

````python
# Remove duplicate rows

df1.drop_duplicates(inplace=True)

print("Dataset Shape After Removing Duplicates:", df1.shape)
````

````python
# Convert date columns into datetime format

date_columns = [
    'Scheduled Delivery Date',
    'Delivered to Client Date',
    'Delivery Recorded Date'
]

for col in date_columns:
    if col in df1.columns:
        df1[col] = pd.to_datetime(df1[col], errors='coerce')
````

````python
# Check updated data types

df1.dtypes
````

````python
# Calculate delivery delay in days

if ('Delivered to Client Date' in df1.columns) and ('Scheduled Delivery Date' in df1.columns):
    df1['Delivery Delay'] = (
        df1['Delivered to Client Date'] -
        df1['Scheduled Delivery Date']
    ).dt.days
````

````python
# Delivery Status

if 'Delivery Delay' in df1.columns:
    df1['Delivery Status'] = df1['Delivery Delay'].apply(
        lambda x: 'Late' if x > 0 else 'On Time'
    )
````
